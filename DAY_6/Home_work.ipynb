```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,500)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = "sine wave",color = "blue",linestyle = "solid")
plt.plot(x,y2,label = "co-sine wave",color = "orange",linestyle = "dashed")
plt.title("sine wave and co sine wave")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.legend()
plt.grid(True)
plt.show()
```
