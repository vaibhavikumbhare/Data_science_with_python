# Introduction to Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10)
y = np.linspace(0,8)

plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x label")
plt.ylabel(" this is y label")
plt.show()
```

```python
a = np.linspace(0,100)
b = np.linspace(0,10)

plt.plot(a,b)
plt.title("sine wave")
plt.xlabel("A label")
plt.ylabel("B label")
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = "sine",color = "red",linestyle = "--")
plt.plot(x,y2,label = "cosine",color = "green",linestyle = "dotted")
plt.title("sine wave and co sine wave")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.legend()
plt.grid(True)
plt.show()
```
