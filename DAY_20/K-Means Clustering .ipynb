# K-Means Clustering 

 what is K-means?
 - k means grps data into clusters based on similarity
 - you can tell the algorithm how many grps (clusters) you want(K)
 - it finds the center of those grps ans asigns data points to the nearst one
 - ex = you have mixst fruits sizes -> K means sperates them into small , medum,large,grps without begin told in advance
   

what is clustering?
- clustering means `grouping similir things together`
- you dont tell the computer the groups,it finds them based on patterns
- real life ex, imagine you have a basket of mixed fruit (apples,orages,bananas) with no labels. the computers looks at size,colour,shape and grp similar fruits same together
  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
data = load_iris()
df = pd.DataFrame(data.data,columns = data.feature_names)
df["species"] = data.target
df.head()
```

```python
# scatter plot
```

```python
sns.scatterplot(x = "petal length (cm)",y = "petal width (cm)",data = df)
plt.title("petal length vs petal width")
plt.show()
```

```python
X = df[["petal length (cm)","petal width (cm)"]]
kmeans = KMeans(n_clusters = 3,random_state = 42)
kmeans.fit(X)
```

```python
df["cluster"]=kmeans.labels_
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = "petal length (cm)",y = "petal width (cm)",data = df,hue="cluster",palette = "deep")
plt.title("petal length vs petal width")
plt.show()
```

# plot cluster center  

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],s = 200,c = "black",marker = "X",label = "Centers")
plt.title("K-means clustering of samples(petal features)")
plt.legend()
plt.show()
```
