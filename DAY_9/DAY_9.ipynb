# Multiple sub-plots

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs = plt.subplots(2)
axs[0].plot(x,y1,color = "red")
axs[0].set_title("sine")
axs[1].plot(x,y2,color = "blue")
axs[1].set_title("co-sine")
plt.tight_layout()
plt.show()
```

# Bar chart

```python
fruites = ["fig","apple","orange","mango"]
prices = [80,90,30,50]
plt.bar(fruites,prices,color = "orange")
plt.title("prices_of_fruits")
plt.xlabel("fruits")
plt.ylabel("prices")
plt.show()
```

# pie char

```python
subjects = ["marathi","english","hindi","histroy","maths","science"]
marks = [80,60,35,90,20,70]
plt.pie(marks,labels = subjects,autopct = "%1.2f%%",startangle = 180)
plt.title("marks_of_subjects")
plt.axis("equal")
plt.show()
```

# Histogram with matplotlib

```python
plt.figure(figsize = (8,5))
plt.hist(marks, color = "blue",edgecolor = "yellow",bins = 5)
plt.title("histogram")
plt.grid(True)
plt.xlabel("marks")
plt.ylabel("number of stud")
plt.show()
```

# Histogram with seaborn

```python
df = pd.DataFrame({"marks":marks})
print(df)
```

```python
plt.figure(figsize = (8,5))
sns.histplot(df["marks"],bins = 5, kde=True, color = "green")
plt.title("histogram with kde")
plt.xlabel("marks")
plt.ylabel("dencity")
plt.show()
```
