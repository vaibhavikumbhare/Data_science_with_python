# 2. Create a function to check if a given no is even or odd using function with arguments 

```python
def even_odd(i):
    if i % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
even_odd(18)
```

# 3. Create a function where we take two arguments first one is for declaration where it starts and second is for where it ends and find the all even no. Between these two arguments and store it in a List and print it. Do it using function with parameter


```python
def even_odd(start,end):
    even_odds = []
    for i in range(start,end+1):
        
        if i%2==0:
            even_odds.append(i)
       
    print(even_odds)
even_odd(1,10)
```

# 4. Create a function where we iterate over a dictionary and print the every element using function without parameter

```python
def dict_elements():
    my_dict = {
        "name": "Vaibhavi",
        "age": 18,
        "city": "Pune",
        "course": "Python"
    }
    
    for key, value in my_dict.items():
        print(key, ":", value)

dict_elements()
```

# Make pie chart of daily activity of your schedule like take one list for activities and other one for hours and make a pie chart of it

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
activities = ["Sleep", "Watching TV", "Scrolling reels", "Eat", "Study","Others"]
hours = [9, 3, 4,3, 1, 5]
plt.pie(hours, labels = activities, autopct="%1.2f%%",startangle = 180)
plt.title("Daily vaibhavi's Schedule")
plt.axis("equal")
plt.show()
```
